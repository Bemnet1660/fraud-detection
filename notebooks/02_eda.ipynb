# 2. Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fraud_df = pd.read_csv('Fraud_Data.csv')
credit_df = pd.read_csv('creditcard.csv')
```

```python
print('Fraud_Data imbalance:\n', fraud_df['class'].value_counts(normalize=True))
print('\nCreditcard imbalance:\n', credit_df['Class'].value_counts(normalize=True))
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,4))
sns.barplot(x=fraud_df['class'].value_counts().index, y=fraud_df['class'].value_counts().values, ax=axes[0])
axes[0].set_title('Fraud_Data Class Distribution')
sns.barplot(x=credit_df['Class'].value_counts().index, y=credit_df['Class'].value_counts().values, ax=axes[1])
axes[1].set_title('Creditcard Class Distribution')
plt.show()
```

```python
sns.boxplot(x='class', y='purchase_value', data=fraud_df)
plt.title('Purchase Value by Class')
plt.show()
```
